```python
import numpy as np

# 벡터 만들기
print("======벡터 만들기======")

# 2차원 벡터 생성 
travel_vector = np.array([300, 400])
print(f"여행자 이동 벡터: {travel_vector}")

# 3차원 벡터 생성
character_position = np.array([2, 3, 5])
print(f"캐릭터 위치: {character_position}")

# 힘 벡터 생성 (물리학 예제)
force_vector = np.array([50, -30, 20])
print(f"작용하는 힘: {force_vector} N")

# 학생 성적 벡터 (5차원)
student_scores = np.array([85, 90, 78, 82, 88])
print(f"학생 성적: {student_scores}")

print("\n=== 벡터의 성분 확인하기 ===")

# 벡터의 각 성분에 접근하기
print(f"캐릭터의 x 좌표: {character_position[0]}")
print(f"캐릭터의 y 좌표: {character_position[1]}")
print(f"캐릭터의 z 좌표: {character_position[2]}")

print(f"학생의 수학 점수: {student_scores[0]}점")
print(f"학생의 영어 점수: {student_scores[1]}점")

print("\n=== 벡터 정보 확인하기 ===")

# 벡터의 차원 확인
print(f"여행자 이동 벡터의 차원: {travel_vector.shape}")
print(f"캐릭터 위치 벡터의 차원: {character_position.shape}")
```

```python
# 벡터의 길이(크기) 계산하기
travel_distance = np.linalg.norm(travel_vector)
print(f"여행자의 이동 거리: {travel_distance:.2f} m")
```

```python
# 실제 계산 예제
print("\n=== 실제 계산해보기 ===")

# 피타고라스 정리로 직선거리 계산
calculated_distance = np.sqrt(travel_vector[0]**2 + travel_vector[1]**2)
print(f"수동 계산한 거리: {calculated_distance:.2f}m")
print(f"NumPy로 계산한 거리: {travel_distance:.2f}m")
```

```python
# 성적 통계
# argmax, argmin는 최대값, 최소값의 인덱스를 반환
subjects = ['수학', '영어', '과학', '사회', '국어']
print(f"\n학생 성적 분석:")
print(f"평균: {np.mean(student_scores):.1f}")
print(f"최고점: {np.max(student_scores)} {subjects[np.argmax(student_scores)]}")
print(f"최저점: {np.min(student_scores)} {subjects[np.argmin(student_scores)]}")
```
