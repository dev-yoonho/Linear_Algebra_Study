```python
# 내적: 두 벡터가 얼마나 비슷한 방향을 향하는지
# 외적: 두 벡터에 수직인 새로운 벡터를 만들어 줌
```

```python
# 두 벡터의 내적은 각 성분의 곱을 모두 더한 것. 즉 결과가 스칼라로 나옴
# 하지만 스칼라 값이 크게 나온다고 해서 강하게 방향이 비슷한게 아니라
# 코사인 유사도를 계산해야만 비로소 비교를 할 수 있음
```

```python
# 중요 공식: 벡터의 내적 = 두 벡터의 노름의 곱 * 두 벡터 사이의 각도의 코사인 값
# 대수적 계산: 벡터의 내적 = 각 성분의 곱을 모두 더한 것
# 기하학적 계산: 벡터의 내적
# = (벡터 a의 크기 * 벡터 a의 x축에 대한 비율) * (벡터 b의 크기 * 벡터 b의 x축에 대한 비율) +
#   (벡터 a의 크기 * 벡터 a의 y축에 대한 비율) * (벡터 b의 크기 * 벡터 b의 y축에 대한 비율))
# = (벡터 a의 크기 * 벡터 b의 크기) * 
#   벡터 a의 x축에 대한 비율 * 벡터 b의 x축에 대한 비율) + 벡터 a의 y축에 대한 비율 * 벡터 b의 y축에 대한 비율)
# = (벡터 a의 크기 * 벡터 b의 크기) * 벡터 a와 벡터 b의 각도 차이 절대값의 x축에 대한 비율
# = 두 벡터의 크기의 곱 * 두 벡터가 이루는 각도의 코사인 값
```

```python
# 각도에 따른 내적 값의 변화
# θ = 0도은 같은 방향이라는 뜻으로 코사인 0도는 1, 즉 내적 값은 두 벡터 크기의 곱
# θ = 90도는 수직이라는 뜻으로 코사인 90도는 0, 즉 내적 값은 0
# θ = 180도는 반대 방향이라는 뜻으로 코사인 180도는 -1, 
# 즉 내적값은 두 벡터 크기의 곱에 -1을 곱한 값
```

```python
# 딥러닝에서의 내적 활용 
# 1. 신경망의 선형 변환
# 신경망의 각 층에서 입력 벡터와 가중치 벡터의 내적을 계산하고 편향 b를 더함.
# 이게 뉴런의 기본 연산
# 내적은 입력 벡터가 가중치 방향에 얼마나 투영되는지 측정하는 연산 
# 신경망이 내적 값을 통해 이 입력이 내가 원하는 패턴인가를 판단
# 뉴런은 특정 패턴을 보고 싶은 감지기
```

```python
# 2. 어텐션 메커니즘
# 트랜스포머에서 쿼리와 키 벡터의 내적으로 어텐션 점수 계산. 내적이 크면 토큰 관련성↑

# 3. 임베딩 유사도
# 단어나 문장의 임베딩 벡터 간의 유사도를 코사인 유사도로 측정
```

```python
# 4. 손실함수
# 오차 측정기 -> 모델의 출력이 얼마나 틀렸는지 숫자로 표현하는 함수
# 모델의 학습은 곧 손실함수 값을 줄이는 방향으로 가중치(파라미터)를 조정하는 과정
```

```python
# 벡터의 직교성
# 벡터가 직교한다는 것은 서로 수직이라는 것. 
# 즉 두 벡터 사이의 각도가 90이며 내적 값이 0
# **두 벡터가 직교할 필요충분조건은 내적이 0인 것이다**

```

```python
# 딥러닝에서의 직교성 활요
# 1. 가중치 초기화
# 직교 벡터들은 서로 독립적인 정보를 표현하므로 
# 가중치를 직교행렬로 초기화하면 안정적 학습 가능
# 가중치를 처음에 랜덤으로 두면 뉴런들이 다 똑같은 특징만 보거나 학습이 안되는 경우 있음
# 뉴런을 독립적으로 출발하게 해 학습 속도를 높이고 정보 손실을 줄이는 효과

# 2. 정규화 기법도 같은 맥락
# 3. 주성분 분석(PCA)
# PCA는 데이터가 가장 넓게 퍼진 방향=가장 정보가 많은 방향을 자동으로 탐색
# 이때 중복 정보를 안 찾기 위해 앞서 찾은 방향과 직교하면서 정보가 많은 방향을 찾는 것
# 노이즈를 제거해 차원 축소를 한다고 생각하자
# 4. 독립성 확보 - 직교하는 벡터 간에는 독립적인 정보를 담고 있으니 표현력도 높아짐
```

```python
# 벡터의 외적
# 외적은 3차원에서만 가능한 연산
# 외적은 두 벡터 모두에 수직인 벡터를 만들어 줌
# 외적의 크기는 두 벡터가 이루는 평행사변형의 넓이와 같으며, 두 벡터의 노름의 곱에 sinθ을 곱함함
```

```python
# 딥러닝에서의 외적 활용
# 1. 컴퓨터 비전 - 3D 모델링
# 외적은 면이 바라보는 방향을 만드는 도구
# 3D 모델링은 모두 삼각형으로 이루어져 있어 그 면의 방향을 찾기 위해 사용

# 2. 로봇 공학
# 로봇은 현재 팔 방향과 목표 팔 방향의 차이를 이용해 회전해야 하는데
# 외적은 그 회전하거나 비틀리는 방향을 계산해줌

# 3. 물리 시뮬레이션
# 토크 계산에 필수. 물체가 얼마나 강하게 어느 방향으로 회전하는지 계산
```

```python
import numpy as np
import math

print("===벡터 내적===")

# 2차원 벡터 내적
a2d = np.array([3, 4])
b2d = np.array([2, 1])
dot_product_2d = np.dot(a2d, b2d)
print(f"a = {a2d}, b = {b2d}")
print(f"내적: a · b = {dot_product_2d}")

# 수동 계산으로 확인
manual_dot = np.sum(a2d * b2d)
print(f"수동 계산 내적: {manual_dot}")
```

```python
# 3차원 벡터 내적
a3d = np.array([1, 2, 3])
b3d = np.array([4, 5, 6])
dot_product_3d = np.dot(a3d, b3d)
print(f"a = {a3d}, b = {b3d}")
print(f"내적: a · b = {dot_product_3d}")
```

```python
print("===벡터 각도 계산===")

# 벡터 사이의 각도
def vector_angle(v1, v2):
    cos_theta = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
    cos_theta = np.clip(cos_theta, -1, 1)
    theta_rad = np.arccos(cos_theta)
    theta_deg = np.degrees(theta_rad)
    return theta_rad, theta_deg

v1 = np.array([1, 0])
v2 = np.array([1, 1])
rad, deg = vector_angle(v1, v2)
print(f"벡터 {v1}와 {v2}의 각도는 {deg:.2f}도")

v3 = np.array([1, 0])
v4 = np.array([0, 1])
rad, deg = vector_angle(v3, v4)
print(f"벡터 {v3}와 {v4}의 각도는 {deg:.2f}도")
```

```python
print("===벡터 직교성 확인===")

# 직교 벡터들
ortho_a = np.array([3, 4])
ortho_b = np.array([4, -3])
dot_ortho = np.dot(ortho_a, ortho_b)
print(f"벡터 {ortho_a}와 {ortho_b}의 내적: {dot_ortho}")
print(f"직교 여부: {abs(dot_ortho) < 1e-10}")
```

```python
# 표준 기저 벡터들
e1 = np.array([1, 0, 0])
e2 = np.array([0, 1, 0])
e3 = np.array([0, 0, 1])
print(f"표준 기저 벡터들:")
print(f"e1 · e2 = {np.dot(e1, e2)}")
print(f"e1 · e3 = {np.dot(e1, e3)}")
print(f"e2 · e3 = {np.dot(e2, e3)}")

print("===벡터 외적===")

# 3차원 벡터 외적
a_cross = np.array([2, 1, 3])
b_cross = np.array([1, 4, 2])
cross_product = np.cross(a_cross, b_cross)
print(f"a = {a_cross}, b = {b_cross}")
print(f"외적: a × b = {cross_product}")
```

```python
# 외적 결과가 원래 벡터들과 수직인지 확인
dot_a_cross = np.dot(a_cross, cross_product)
dot_b_cross = np.dot(b_cross, cross_product)
print(f"a · (a × b) = {dot_a_cross} (0에 가까워야 함)")
print(f"b · (a × b) = {dot_b_cross} (0에 가까워야 함)")
```

```python
# 외적의 크기와 평행사변형 넓이
cross_magnitude = np.linalg.norm(cross_product)
print(f"외적의 크기: {cross_magnitude:.2f}")
```

```python
# 예제 1: 물리학 - 일 계산
force = np.array([50, 30])
displacement = np.array([4, 2])
work = np.dot(force, displacement)
print(f"힘: {force} N, 변위: {displacement} m")
print(f"일: {work} J")

# 예제 2: 코사인 유사도
doc_a = np.array([3, 4, 0])
doc_b = np.array([1, 2, 2])
cosine_similarity = np.dot(doc_a, doc_b) / (np.linalg.norm(doc_a) * np.linalg.norm(doc_b))
print(f"문서 a: {doc_a}, 문서 b: {doc_b}")
print(f"문서 a와 b의 코사인 유사도: {cosine_similarity:.3f}")

# 예제 3: 게임 - 시야 각도 체크
view_direction = np.array([1, 0])
enemy_position = np.array([3, 4])
enemy_direction = enemy_position / np.linalg.norm(enemy_position)

dot_view_enemy = np.dot(view_direction, enemy_direction)
angle_to_enemy = np.degrees(np.arccos(np.clip(dot_view_enemy, -1, 1)))

print(f"시야 방향: {view_direction}, 적 위치: {enemy_position}")
print(f"적까지의 각도: {angle_to_enemy:.2f}도")

fov = 60

if angle_to_enemy <= fov:
    print(f"적이 시야({fov}도)안에 있습니다.")
else:
    print(f"적이 시야({fov}도) 밖에 있습니다.")
```

```python
# 예제 4: 법선 벡터
edge1 = np.array([1, 0, 0])
edge2 = np.array([0, 1, 0])
normal = np.cross(edge1, edge2)
print(f"\n삼각형 변1: {edge1}")
print(f"삼각형 변2: {edge2}")
print(f"법선 벡터: {normal}")
```

```python
# 교환법칙
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(f"교환법칙: a·b = {np.dot(a, b)}, b·a = {np.dot(b, a)}")

# 분배법칙
c = np.array([7, 8, 9])
left_side = np.dot(a, b + c)
right_side = np.dot(a, b) + np.dot(a, c)
print(f"분배법칙: a·(b+c) = {left_side}, a·b + a·c = {right_side}")

# 자기 자신과의 내적
self_dot = np.dot(a, a)
norm_squared = np.linalg.norm(a) ** 2
print(f"a·a = {self_dot}, ||a||² = {norm_squared:.6f}")
```
